**Tratamento dos dados**

Objetivo: Realizar a abertura dos arquivos e ler apenas as variáveis de interesse dos quinze
primeiros colocados das copas de cada arquivo. A lista de valores de cada variável devem
ser a junção de todos os valores contidos daquela variável nos 3 CSVs.

**Exemplo:**
FIFA - 2006.csv, Goals For = 6, 4, 1, 2, 8.
FIFA - 2010.csv, Goals For = 1, 3, 4, 5, 6.
FIFA - 2014.csv, Goals For = 5, 4, 2, 3, 6.
FIFA - 2018.csv, Goals For = 9, 8, 4, 5, 7.
Lista de valores da variável “Goals For”: 6, 4, 1, 2, 8, 1, 3, 4, 5, 6, 5, 4, 2, 3, 6, 9, 8, 4,
5, 7.

**Análise estatística**

Objetivo: Geração de média, desvio padrão e limites superiores e inferiores de cada variável
de interesse. Nesse projeto, o limite superior é dado a partir da soma de 3 desvios padrão à
média e o limite inferior é dado a partir da subtração de 3 desvios padrão da média.

**Exemplo:**
Média: 6 / Desvio padrão: 0,5

Limite superior: 6 + 3*(0,5) = 7,5 / Limite inferior: 6 - 3*(0,5) = 4,5


**Plotagem de gráficos**

Objetivo: Gerar um gráfico de linha para cada variável escolhida. Esses gráficos devem ter,
além de uma linha para os valores da variável, uma linha para o limite inferior e uma para o
limite superior. Ou seja, cada gráfico terá 3 linhas.

**Parte 1 -** Importando Bibliotecas, Numpy para realizar as tarefas matemáticas necessárias ao problema (média e desvio padrão); Pandas para fazer o tratamento de dados provenientes dos CSVs; Matplotlib.pyplot para a plotagem e visualização gráfica dos dados

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Parte 2 -** Fazendo a leitura dos dados contidos nos CSVs

```python
copa2006 = pd.read_csv(r'fifa2006.csv', usecols = ['Team', 'Goals For', 'Goals Against'], index_col=0, nrows=15)
copa2010 = pd.read_csv(r'fifa2010.csv', usecols = ['Team', 'Goals For', 'Goals Against'], index_col=0, nrows=15)
copa2014 = pd.read_csv(r'fifa2014.csv', usecols = ['Team', 'Goals For', 'Goals Against'], index_col=0, nrows=15)
copa2018 = pd.read_csv(r'fifa2018.csv', usecols = ['Team', 'Goals For', 'Goals Against'], index_col=0, nrows=15)
```

**Parte 4 -** Transformando os dados de cada ano em listas próprias e exibindo para verificar

```python
goals_for2006 = copa2006['Goals For'].tolist()
goals_against2006 = copa2006['Goals Against'].tolist()
print('Gols a Favor 2006:', goals_for2006)
print('Gols Contra 2006:', goals_against2006)

goals_for2010 = copa2010['Goals For'].tolist()
goals_against2010 = copa2010['Goals Against'].tolist()
print('Gols a Favor 2010:', goals_for2010)
print('Gols Contra 2010:', goals_against2010)

goals_for2014 = copa2014['Goals For'].tolist()
goals_against2014 = copa2014['Goals Against'].tolist()
print('Gols a Favor 2014:', goals_for2014)
print('Gols Contra 2014:', goals_against2014)

goals_for2018 = copa2018['Goals For'].tolist()
goals_against2018 = copa2018['Goals Against'].tolist()
print('Gols a Favor 2018:', goals_for2018)
print('Gols Contra 2018:', goals_against2018)
```

**Parte 5 -** Criando uma lista com todos os valores de gols a favor dos 4 anos e exibindo-a

```python
goals_for = goals_for2006 + goals_for2010 + goals_for2014 + goals_for2018
print('Gols a Favor Total:', goals_for)
```

**Parte 6 -** Criando uma lista com todos os valores de gols contras dos 4 anos e exibindo-a

```python
goals_against = goals_against2006 + goals_against2010 + goals_against2014 + goals_against2018
print('Gols Contra Total:', goals_against)
```

**Parte 7 -** Fazendo as operações estatístico-matemáticas necessárias para os gols a favor

```python
mean_goals_for = np.mean(goals_for)
standard_deviation_goals_for = np.std(goals_for)
superior_limit_goals_for = mean_goals_for + (3*standard_deviation_goals_for)
inferior_limit_goals_for = mean_goals_for - (3*standard_deviation_goals_for)
print('Média Gols a Favor:', mean_goals_for)
print('Desvio Padrão Gols a Favor:', round(standard_deviation_goals_for, 2))
print('Limite Superior Gols a Favor:', round(superior_limit_goals_for, 2))
print('Limite Inferior Gols a Favor:', round(inferior_limit_goals_for, 2))
```

**Parte 8 -** Fazendo as operações estatístico-matemáticas necessárias para os gols contras

```python
mean_goals_against = np.mean(goals_against)
standard_deviation_goals_against = np.std(goals_against)
superior_limit_goals_against = mean_goals_against + (3*standard_deviation_goals_against)
inferior_limit_goals_against = mean_goals_against - (3*standard_deviation_goals_against)
print('Média Gols Contra:', mean_goals_against)
print('Desvio Padrão Gols Contra:', round(standard_deviation_goals_against, 2))
print('Limite Superior Gols Contra:', round(superior_limit_goals_against, 2))
print('Limite Inferior Gols Contra:', round(inferior_limit_goals_against, 2))
```

**Parte 9 -** Plotando gráfico dos gols a favor

```python
eixoX = list()
superior = list()
inferior = list()
for i in range (0, 60):
  eixoX.append(i)
  superior.append(superior_limit_goals_for)
  inferior.append(inferior_limit_goals_for)
plt.figure (figsize = (12, 7)) 
plt.title('Gols a Favor (Copas de 2006, 2010, 2014 e 2018)', fontsize = 16)
plt.xlabel('Quantidade de Dados',fontsize = 12)
plt.ylabel('Quantidade de Gols',fontsize = 12)
plt.plot(eixoX, superior, label='Limite Superior', c = 'blue')
plt.plot(eixoX, goals_for, label='Gols a Favor', c = 'green')
plt.plot(eixoX, inferior, label='Limite Inferior', c = 'gray')
plt.ylim(-7,26)
plt.legend(loc = 'upper right')
plt.show()
```

**Parte 10 -** Plotando gráfico dos gols contras

```python
eixoX = list()
superior = list()
inferior = list()
for i in range (0, 60):
  eixoX.append(i)
  superior.append(superior_limit_goals_against)
  inferior.append(inferior_limit_goals_against)
plt.figure (figsize = (12, 7)) 
plt.title('Gols Contra (Copas de 2006, 2010, 2014 e 2018)', fontsize = 16)
plt.xlabel('Quantidade de Dados',fontsize = 12)
plt.ylabel('Quantidade de Gols',fontsize = 12)
plt.plot(eixoX, superior, label='Limite Superior', c = 'blue')
plt.plot(eixoX, goals_against, label='Gols Contra', c = 'red')
plt.plot(eixoX, inferior, label='Limite Inferior', c = 'gray')
plt.ylim(-6,16)
plt.legend(loc = 'upper right')
plt.show()
```
